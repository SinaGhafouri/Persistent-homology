## <a href="https://ripser.scikit-tda.org/en/latest/">Documentation page</a>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# for working with point clouds, ripser is much faster and more efficient that gudhi.
from ripser import ripser
from persim import plot_diagrams
```

## Random 2D point cloud

```python
N = 100 # size
x = np.random.random(N)
y = np.random.random(N)
plt.plot(x,y,'.');
```

## Computing simplicial complex persistence:

```python
points = np.vstack((x,y)).T

dgms = ripser(points, maxdim=1)['dgms']
H0, H1 = dgms
```

```python
plot_diagrams(dgms)
```
